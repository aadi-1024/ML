```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv')
# print(train.head())

train_data = train[["Pclass", "Sex", "Age", "SibSp", "Fare"]]

cols = train_data.columns

train_result = train["Survived"]
```

```python
for i in cols:
    plt.hist(train[train['Survived'] == 1][i], color='blue', alpha=0.7)
    plt.hist(train[train['Survived'] == 0][i], color='red', alpha=0.7)
    plt.xlabel(i)
    plt.show()
```

```python
train_data.loc[train_data['Sex'] == 'male'] = 1
```

```python
train_data.loc[train_data['Sex'] == 'female'] = 0
```

```python
from sklearn import neighbors
```

```python
knn = neighbors.KNeighborsClassifier()
```

```python
knn.fit(train_data, train_result)
```

```python
test = pd.read_csv('test.csv')
test_data = test[cols]
test_data.loc[test_data['Sex'] == 'male'] = 1
test_data.loc[test_data['Sex'] == 'female'] = 0
res = knn.predict(test_data)
```

```python
res = pd.DataFrame(res, columns=["Survived"])
```

```python
pd.concat([test['PassengerId'], res['Survived']])
```
